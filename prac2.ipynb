```python
#################################################################
########## 단순회귀는 인과성이 보장된 것만 써야한다! ############
#################################################################
# 과학적 근거가 있어야 한다. (A라는 물질이 B를 일으킨다)
# 대조군/실험군이 확실할 때 (한 쪽은 A를 섭취, 다른 한쪽은 A라는 물질을 비섭취)
# 윤리적일 때

#### 상관관계, 인과성을 확일할 떄 주의할 것 ####
# 원인과 격롸가 바뀌었을 가능성
# 우연

A = np.array([1,1,1,1,1,1])  # 한쪽이 상수이면 상관관계는 0
B = np.array([1,3,5,7,9,11])

cov = np.cov(A,B)
cov
```

```python
# 회귀(regression) - 세상엔 완벽한게 있고 + 지저분한게 있어서, 완벽한 것을 찾는 것
# y = ax + b + e 에서 y = ax + b 를 찾자
# COV(X, Y) = COV(x, ax+b+e(오류,입실론))
# COV(X, aX+b+e) = COV(X,aX) + COV(X,b) + COV(X, e)
#                = a COV(X,X) + COV(X, b) + COV(X, e)
#                = a V(x)
#              a = COV(X,Y) / V(X)
#              Y = aX + b + e
#           e(Y) = E(aX) + E(b) + E(e)
#                = aE(X) + b
#              b = E(Y) - aE(X)
#                = E(Y) - ((COV(X,Y)/V(X))*E(X))

# y = a x + b --> 1차 = 단순회귀
#     a x^2   --> 2차 = 2차회귀
#             --> 3차 = 3차회귀
# 1~2차에서 끝내는게 좋음

# y = ax + b, x:appearance, y:personality
a = np.cov(ap, pe)[0,1]/np.cov(ap)
b = np.mean(pe)-a*np.mean(ap)
a, b
```

```python
# factorial
# combination
#                         a C b = a! / b! (a-b)!
# 6개에서 3개를 뽑을 확률 6 C 3 = 6! / 3! (6-3)! = 6! / 3! * (6-3)! = 6! / 3! * 3! = 6*5*4 / 3!
# 3개에서 1개를 뽑을 확률 3 C 1
# 4개에서 2개를 뽑을 확률 4 C 2

# 자연대수(자연로그), exponential, e 대략 2.7
# lim(x->0) (1+x)^(1/x)

# 이항 분포 - n번의 독립 시행, 각 확률이 p일 때, 이산 확률 분포
# ex) 40%로 100번 돌리면 몇 번 이길까?

# 포아송 분포 - 단위 시간 안에 어떤 사건이 몇 번 발생할지 확률 분포
# ex) 일어날 확률이 굉장히 적을 때
# ex) 범죄, 사고, 화재 -----> 이항분포로 돌리게되면 ##과장##될 확률이 높다!

# P = n C k (p)^k * (1-p)^(n-k)

# 남,여 에서 남자가 태어날 확률 0.52
#            k명이 태어날 확률 = P = 100 C k (0.52)^k * (0.48)^(100 - k)

# 평균 = n (횟수) * p (확률) .....
# 분산 = n p (1 - p) 
```

```python
from math import factorial as fact 
import math 

# 이항 분포 a C b = a! / b! (a-b)!
#             P = n C k (p)^k * (1-p)^(n-k)
def bin_prob(n,k,p):
  nck = fact(n)/(fact(k)*fact(n-k))
  exp = nck * math.pow(p,k) * math.pow((1-p),(n-k))
  return exp 

bin_prob(100,50,0.52) * 100 #100명의 아이중에 남자 50명이 나올 확률, 남자가 나올 확률(0.52)
```

```python
bin_prob(120,20,1/6) * 100 # 주사위를 120번 돌렸는데 한 번호가 20번 나올 확률, 한 번호가 나올 확률은 1/6
```

```python
# 정규분포 

expect_ = []
n = 100
p = 0.52

for i in range(n):
  expect_.append(bin_prob(n,i,p)*100)

plt.bar(np.arange(n), expect_)
```

```python
# 포아송 분포
# n -> 무한대
# k -> 0 으로 수렴

# 일어날 횟수 - λ(람다) 번일 때
# 이 사건이 n번 일어날 확률

# P(n) = (λ^n * e^-λ) / n!
# np       E = λ
# np(1-p)  V = λ

from math import exp 

def poi_prob(n, lammda):
  exp = (math.pow(lammda, n) * math.exp(-lammda)) / fact(n)
  return exp 

expect_ = []

total = 40 # 갯수
lammda = 2 # 사건이 두 번 일어남

for i in range(total):
  expect_.append(poi_prob(i, lammda)*100)

plt.bar(np.arange(total), expect_)
# 이렇게 이항분포와 다르게 낮은확률로 일어나는 사건들은 포아송 분포로 표시하는것이 좋다
# 이항분포로 표시할시 정규분포형태를 따르기때문에 위험도가 과장될 수가 있다.
```

```python
# 포아송 분포 - 우연인지 아닌지
# t-test      - 실험 전후를 비교 (ex: 약먹고 난후, 치료법 전후, 2개 비교할 떄) : 보통 paired t-test를 씀
# anova test  - 다른 집단 비교 (특정 조건은 없다. n개 이상 비교할 때, ex: 한국인, 필리핀인, 미국인의 키 비교) 
#               Basically, you're testing groups to see if there's a difference between them.

from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

breast_cancer = load_breast_cancer()

df = pd.DataFrame(data = breast_cancer.data, columns = breast_cancer.feature_names)

df.head()
```

```python
mean_radius = df["mean radius"]
worst_radius = df["worst radius"]

worst_radius, mean_radius
```

```python
mean_radius = np.array(mean_radius)
worst_radius = np.array(worst_radius)

worst_radius, mean_radius
```

```python
plt.scatter(mean_radius, worst_radius)
```

```python
#### 상관관계가 높다고해서
#### 1. 인과관계가 아니다
#### 2. 우연일 가능성이 있다 --> chi2 제곱

corr = np.corrcoef(mean_radius, worst_radius)
corr[0,1]
```

```python
# 카이제곱 검정 - 두 범주형 변수가 서로 상관이 있는지 독립인지를 판단하는 통계적 검정방법을 
#                 교차분석 또는 카이제곱검정이라 한다
#               - 주로 문자 데이터(범주형 변수) 간의 연관성을 검증하는 데 활용된다.
#               - 과학적 근거, 연관이 있다면 카이제곱 검정을 안해도 된다.
#               - 상관관계가 높다고해도 p가 낮을 수도 있다. 외부요인이 존재하기 떄문

from scipy.stats import chi2_contingency

data = pd.DataFrame([mean_radius, worst_radius])
A, p, dof, value = chi2_contingency(data) # p가 중요
p                                      #### p < 0.05 서로 영향을 준다 = 우연이 아니다
                                          # p > 0.05 우연일 가능성이 높다
```

```python
# 포아송 분포

# 일어날 횟수 - λ(람다) 번일 때
# 이 사건이 n번 일어날 확률

from math import exp
from math import factorial as fact

def poi_prob(n, lammbda):
  expect = (pow(lammbda,n) * exp(-lammbda)) / fact(n)
  return expect

expect_ = []

total = 100 # 갯수
lammbda = 5 # 사건이 두 번 일어남

for i in range(total):
  expect_.append(poi_prob(i, lammda)*100)

plt.bar(np.arange(total), expect_)
```

```python
# T-Test
# A집단과 B집단이 다르다는 것을 테스트
# p < 0.05 보다 작으면 A집단과 B집단은 다르다
# ex) 남자와 여자, 20대와 30대, 인간과 원숭이 등.... 

# 독립변수 T-Test (independent t-test)
# 대응변수 T-Test (paried t-test) ---- 절대적으로 중요!
# 1. 표본 수 동일  - 즉 같은 조직
# 2. 동분산성 보장 - 즉 같은 조직
# ex) 실험, 자극, 약, 사건 전후........... ㅁㅁ했더니 달라졌다, 효과가 있다,변했다 등

# 독립변수 T-Test
import scipy.stats as stats

A = [169,170,171,172,168,180,171,167] # 연구실 아이들
A = np.array(A)

B = [180,175,168,187,190,191,195,189,178]# 연구실 외국아이들
B = np.array(B)

C = [168,170,169,172,173,172,166,173] # A의 연구실아이들 -> 노마 먹고 키 큼
C = np.array(C)

D = [183,185,190,195,186,186,178,195] # A의 연구들이 -> 노마2 먹고 키 엄청 큼
D = np.array(D)

t_ind, p = stats.ttest_ind(A,B)
p              ### p < 0.05 --> 유의미한 차이가 있다
```

```python
# paired t-test 서로 갯수 맞아야 함
t_rel, p =stats.ttest_rel(A,C)
p              ### p > 0.05 --> 차이가 없다!
```

```python
t_rel, p = stats.ttest_rel(A,D)
p              ### p < 0.05 --> 유의미한 차이가 있다!
```
